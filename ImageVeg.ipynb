```python
from bs4 import BeautifulSoup
import json 
import requests
```

```python
#URL = "http://www.ngchina.com.cn/animals/"
URL = "https://www.instacart.ca/t-t/aisles/63205-fresh-vegetables/page/1"
#hdrs = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36'}
hdrs = {'User-Agent':'Chrome/78.0.3904.70'}
html = requests.get(URL,headers=hdrs).text
soup = BeautifulSoup(html, 'lxml')
#img_ul = soup.find_all('ul', {"class": "img_list"})
img_ul = soup.find_all('script', {"type":"application/ld+json"})
import os
os.makedirs('./img/', exist_ok=True)
x = 0
for img in img_ul[1:]:
    json_str = json.loads(img.text)
    image_url = json_str['image'].replace('primary','large')
    image_name = json_str['name'] + '.jpg'
    print(image_url,type(image_url))
    print(image_name)
    x = x + 1
    r = requests.get(image_url, stream=True)
    with open('./img/%s' % image_name, 'wb') as f:
        for chunk in r.iter_content(chunk_size=128):
            f.write(chunk)
    print('Saved %s' % image_name)
    print(f'--------------------------------------Image Scraped ({x}) -----------------------------------')
#soup1 = BeautifulSoup(img_ul,'lxml')
#js_data = json.loads(soup.find_all('div',{"type":"application/ld+json"}))
#for ul in img_ul:
#    print(ul.find_all('image'))
#    for img in ul.find_all('img'):
#        print(img)
#        url = img['src']
#        print(url)
#img_ul.find_all('img')
#js_data = json.loads(img_ul)
#js_data
```
